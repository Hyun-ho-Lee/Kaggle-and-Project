```python
cat_var = ['gender', 'reality','car', 'income_type', 'edu_type', 'family_type', 'house_type','FLAG_MOBIL','work_phone','phone','email','occyp_type']

def plot_cat(df):
    for feature in cat_var:
        if feature == 'occyp_type':
            plt.figure(figsize = (15,12))
            sns.catplot(x = train[feature], y = train['credit'], data = train, kind = 'bar')
            plt.title('Count of ' + str(feature), size = 15)
            plt.xlabel(feature, size = 12)
            plt.xticks(rotation = 90)
            plt.show()
            
        else:
            plt.figure(figsize = (6,6))
            sns.catplot(x = train[feature], y = train['credit'], data = train, kind = 'bar')
            plt.title('Count of ' + str(feature), size = 15)
            plt.xlabel(feature, size = 12)
            plt.xticks(rotation = 45)
            plt.show()

plot_cat(train)
```
